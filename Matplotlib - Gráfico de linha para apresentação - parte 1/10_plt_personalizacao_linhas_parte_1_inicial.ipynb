## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt  
import numpy as np  

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']  
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]  

plt.figure(figsize=(10, 6))  
plt.plot(meses, vendas, marker='o', color='#007ACC', linestyle='-', linewidth=2, markersize=6, label='Vendas')  
plt.xlabel('Mês', fontsize=12)  
plt.ylabel('Vendas (em milhares)', fontsize=12)  
plt.title('Crescimento das Vendas ao Longo do Ano', fontsize=14, fontweight='bold')  
plt.grid(visible=True, linestyle='--', alpha=0.6)  

max_venda = max(vendas)  
max_index = vendas.index(max_venda)  
plt.annotate(f'Máximo: {max_venda} mil', xy=(meses[max_index], max_venda), xytext=(meses[max_index], max_venda-2), arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='black')
min_venda = min(vendas)  
min_index = vendas.index(min_venda)  

plt.plot(meses[min_index], min_venda, 'rx', markersize=10, label='Valor Mínimo')  
plt.annotate(f'Mínimo: {min_venda} mil', xy=(meses[min_index], min_venda), xytext=(meses[min_index-1], min_venda-5), arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=10, color='red')

inicio_sombreamento = 6  
plt.axvspan(inicio_sombreamento, len(meses), alpha=0.3, color='green')  
plt.text(8, 85, 'Recuperação\nde crescimento', fontsize=10, color='darkgreen', ha='center') 

plt.legend()  
plt.tight_layout()  
plt.show()
```
