## Gráfico de barras empilhadas para mostrar a distribuição de diferentes tipos de vendas ao longo do tempo

Considere que você tenha uma empresa e queira mostrar a distribuição de diferentes tipos de vendas ao longo do tempo. Crie um gráfico de barras empilhadas:

```python
import matplotlib.pyplot as plt

anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20] # em milhões de reais
vendas_loja = [10, 8, 6, 4] # em milhões de reais
```

```python
cores = plt.get_cmap('Set3').colors
plt.bar(anos, vendas_online, label='Vendas Online', color=cores[0])
plt.bar(anos,vendas_loja, label='Loja física', bottom=vendas_online, color=cores[1] ) # para barras empilhadas
plt.legend()
plt.xticks(ticks=anos, labels=['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4']) # pode ser usado apenas ticks=anos, para pegar os valor da lista anos
plt.ylabel('Vendas em milhões de reais')
plt.title('Vendas Online vs Loja Física')
plt.show()
```

```python
#utilizando Cycler
from cycler import cycler
cores_cycler = cycler('color', plt.get_cmap('Set3').colors)
plt.rc('axes', prop_cycle=cores_cycler)
plt.bar(anos, vendas_online, label='Vendas Online')
plt.bar(anos,vendas_loja, label='Loja física', bottom=vendas_online) # para barras empilhadas
plt.legend()
plt.xticks(ticks=anos, labels=['Ano 1', 'Ano 2', 'Ano 3', 'Ano 4']) # pode ser usado apenas ticks=anos, para pegar os valor da lista anos
plt.ylabel('Vendas em milhões de reais')
plt.title('Vendas Online vs Loja Física')
plt.show()
```

```python
# a partir do plt.rc todos os gráficos terão as cores da paleta definida no cycler
anos = [2021, 2022, 2023, 2024]
vendas = [5, 10, 15, 20] # em milhões de reais

plt.plot(anos, vendas, label='Vendas')
plt.show()
```
